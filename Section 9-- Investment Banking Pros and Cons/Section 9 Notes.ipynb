### Lect 42: 7 Interview Tips
Part 1: Quantitative Tips

<b> Q1) "Company X is considering acquiring Company Y. How would you analyze this situation?

A1)
- Is Company Y a competitor or in a different market?
    - If Competitor:
        - talk about expense, headcount synergies
        - Keep drilling down until you get an EPS number that's accretive or dillutive
    - If Different Market:
        - Talk about revenue synergies
        - Use existing distribution channels to increase revenue for newly acquired company
        - Keep drilling down until you get an EPS number that's accretive or dillutive

Q2) Company X is considering entering a market by buying or building. What would you advise?
    
A2) Four questions to ask
    1) Total addressable market?
    
    2) % of market you're looking to capture?
    
    3) Cost to build?
    
    4) Cost to buy?
    
<b> BUILDING:
    
    - Is building really cost effective? To we have the talent to build a true competitor?
    - How many years would it take to become credible/competitive?
    - What happens if the building fails?
    - Impact on short/long term EPS?
    
<b> BUYING:
    
    - How urgent is this? How quickly do we need to get to market?
    - What happens if the deal doesn't work?
    - Impacting on short/long term EPS?
    

Q3) Company X is considering an IPO. How would you aproach the situation?
    
A3)
    
    Start by asking what the valuation is
        GO PUBLIC:
            - Competition's P/E, peg. P/R are more expensive
            - The IPO market is robust 
            - If recent companies in the same sector had successful IPOs
        DON'T GO PUBLIC:
            - If the competitiont's valuation is cheaper (with similar growth)
            - The IPO market is awful
            - If recent companies in the same sector had awful IPOs
    BONUS:
    - Maybe you can test the market by conducting a Non Deal Roadshow


Q4) Explain how to build a financial model from scratch

A4) 
    1) Is this a public or private company?
        
<b> Public:
    
    - Get Model data from sec.gov from the past 10 years using 10-Ks or S-1s (if recent IPO) and/or visit Invenstors Relations portion of website
    - Create Income Statement, Balance Sheet, Cash Flow Statement 
    - Analyze past trends and forecast the next 5 fiscal years with a large emphasis of Revenue
    - Is it possible to speak to the CFO to better understand if my model assumptions were correct or not 
   
<b> Private:
   
    - Can I ask the company for Financials?
    - Find out what the TAM (Total Addressable Market) is
        - If you can't, add up the revenue from the top 25 firms as an estimate
    - Decide what percent of the TAM the company will have in 10 years
    - Project sales aand make most Income Statement elements a % of revenue
